```python
idade = int(input('Digite sua idade: '))
#SE idade menor que 18, executa o print
if idade < 18:
  print('Acesso bloqueado') #repare no espaço(TABulação) antes.
print('Fim do programa') #executa sempre
```

Chamamos de identação o espaçamento de cada linha de código. A identação marca quais comando fazem parte do IF

```python
#print associado ao IF
if idade < 18:
  print('Acesso bloqueado')

#ERRO falta identação
if idade< 18:
print('Acesso bloqueado')

#Dois prints associadas ao if, terceiro print, independente
if idade<18:
  print('Acesso')
  print('Bloqueado')
print('Fim')
```

#Funções

Funções são uma forma de dar apelidos para trechos de código. Serve para melhorar sua organização e diminuir a reescrita de trechos de códigos similares

```python
print('|',"_"*18,"|")
print('|',"_"*18,"|")
print('          MENU')
print('|',"_"*18,"|")
print('|',"_"*18,"|")
```

```python
def darDestaque() :
  print('||',"_"*18,"||")
  print('||',"_"*18,"||")


darDestaque()
print('          MENU')
darDestaque()
```

Funções com parâmetro

```python
def darDestaque(mensagem) :
  print('||',"_"*18,"||")
  print('||',"_"*18,"||")
  print('          ',mensagem)
  print('||',"_"*18,"||")
  print('||',"_"*18,"||")


darDestaque('Olá')
darDestaque('Menu')
```

```python
def potencia (base,expoente) :
  resp=1
  #multiplicacao da base tantas vezes quanto o expoente
  for i in range(expoente):
    resp *= base  # resp = resp * base
  return resp


print(potencia(2,10))
```

>{'cidade' : 'Curitiba', 'latitude':-25.2548, 'longitude':-49.1615}

Nesse momento descubra qual a cidade mais ao norte, sul, leste e oeste.


```python
#listas vazias que vão armazenar as informações
listaCidades= []
cidadeCoordenada = {}

#leitura da lista de cidades
nome = input('Digite uma cidade: ')
while nome != '' :
  cidadeCoordenada['cidade'] = nome
  cidadeCoordenada['latitude'] = float(input('Digite uma latitude: '))
  cidadeCoordenada['longitude'] = float(input('Digite uma longitude: '))
  listaCidades.append(cidadeCoordenada.copy())
  nome = input('Digite uma cidade: ')

#colocando em ordem crescente por latitude e longitude
listaNorteSul =   sorted(listaCidades,key=lambda dicionario: dicionario['latitude'])
listaLesteOeste = sorted(listaCidades,key=lambda dicionario: dicionario['longitude'])

#pegando o primeiro e o último elemento da lista para saber qual é a menor e maior longitude e latitude
norte = listaNorteSul[-1]
sul =   listaNorteSul[0]
leste = listaLesteOeste[-1]
oeste = listaLesteOeste[0]

print(f'A cidade mais ao norte é {norte["cidade"]} ')
print(f'A cidade mais ao sul é {sul["cidade"]} ')
print(f'A cidade mais ao leste é {leste["cidade"]} ')
print(f'A cidade mais ao oeste é {oeste["cidade"]} ')
```
